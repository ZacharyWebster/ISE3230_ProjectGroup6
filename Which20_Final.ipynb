```python
import gurobipy as gp
from gurobipy import GRB

m = gp.Model("Phase 2")
m.Params.LogToConsole = 1  # use this to disable detailed result output

x = m.addVars(60, vtype=GRB.BINARY, name="x")

facilityID = []
with open (r"C:\Users\joshc\OneDrive\Desktop\AU 2025\ISE Project\workArea14FacilityID.txt") as fID:
    for line in fID:
        facilityID.append(line.strip())

priority = []
with open (r"C:\Users\joshc\OneDrive\Desktop\AU 2025\ISE Project\workArea14Priority.txt") as pri:
    for line in pri:
        priority.append(int(line.strip()))

inAppeal = []
with open (r"C:\Users\joshc\OneDrive\Desktop\AU 2025\ISE Project\workArea14InAppeal.txt") as inApp:
    for line in inApp:
        inAppeal.append(int(line.strip()))

numBuys = []
with open(r"C:\Users\joshc\OneDrive\Desktop\AU 2025\ISE Project\workArea14numBuys.txt") as numBu:
    for line in numBu:
        numBuys.append(int(line.strip()))

been14Days = []
with open(r"C:\Users\joshc\OneDrive\Desktop\AU 2025\ISE Project\workArea1414Days.txt") as days:
    for line in days:
        been14Days.append(int(line.strip()))

hadAC = []
with open(r"C:\Users\joshc\OneDrive\Desktop\AU 2025\ISE Project\workArea14hadAC.txt") as AC:
    for line in AC:
        hadAC.append(int(line.strip()))

obj = 0
for i in range(60):
    obj += priority[i] * x[i]
    
m.setObjective(obj, GRB.MAXIMIZE)

constr = 0
for i in range(60):
    constr += x[i]
    
m.addConstr(constr <= 20)

for i in range(60):
    m.addConstr(x[i] + inAppeal[i] <= 1)

for i in range(60):
    m.addConstr(x[i] + numBuys[i] <= 3)

for i in range(60):
    m.addConstr(x[i] + been14Days[i] <= 1)

for i in range(60):
    m.addConstr(x[i] + hadAC[i] <= 1)

m.optimize()

print("obj_func = ", m.objVal)

for i in range(60):
    if x[i].x == 1:
        print('Visit facility ID: %s' % (facilityID[i]))
```
